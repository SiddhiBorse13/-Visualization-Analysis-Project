## Cleaning data

### Remove unwanted columns from dataset

#### Several columns are being dropped from the dataset using the drop function with the inplace=True parameter.The columns being dropped are 'rank','Abbreviation','country_rank','channel_type_rank','video_views_for_the_last_30_days','lowest_monthly_earnings','highest_monthly_earnings','subscribers_for_last_30_days','created_date','Gross tertiary education enrollment(%)','Unemployment rate','Urban_population','Latitude',and 'Longitude'. This operation effectively removes these columns from the dataset.

```python
df.drop(['rank'],axis=1,inplace=True)
df.drop(['Abbreviation'],axis=1,inplace=True)
df.drop(['country_rank'],axis=1,inplace=True)
df.drop(['channel_type_rank'],axis=1,inplace=True)
df.drop(['video_views_for_the_last_30_days'],axis=1,inplace=True)
df.drop(['lowest_monthly_earnings'],axis=1,inplace=True)
df.drop(['highest_monthly_earnings'],axis=1,inplace=True)
df.drop(['subscribers_for_last_30_days'],axis=1,inplace=True)
df.drop(['created_date'],axis=1,inplace=True)
df.drop(['Gross tertiary education enrollment (%)'],axis=1,inplace=True)
df.drop(['Unemployment rate'],axis=1,inplace=True)
df.drop(['Urban_population'],axis=1,inplace=True)
df.drop(['Latitude'],axis=1,inplace=True)
df.drop(['Longitude'],axis=1,inplace=True)
```
